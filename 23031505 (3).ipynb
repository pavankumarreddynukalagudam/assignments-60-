```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting plot styles
sns.set(style="whitegrid")

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data preparation for Match Outcomes by Venue
venue_wins = df_matches.groupby(['venue', 'winner']).size().unstack(fill_value=0)

# Plotting Match Outcomes by Venue
plt.figure(figsize=(18, 12))
sns.heatmap(venue_wins, annot=True, fmt="d", cmap="Blues")
plt.title('Match Outcomes by Venue')
plt.xlabel('Teams')
plt.ylabel('Venue')
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
# Saving the formatted graph as a PNG file
plt.savefig('chart1.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Data preparation for Top Players by Player of Match Awards
top_players = df_matches['player_of_match'].value_counts().head(10)

# Plotting Top Players by Player of Match Awards
plt.figure(figsize=(12, 6))
sns.barplot(x=top_players.index, y=top_players.values, palette='viridis')
plt.title('Top Players by Player of Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=45)
plt.tight_layout()
# Saving the formatted chart as a PNG file
plt.savefig('chart2.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Summing up runs scored by each player
player_runs = df_ball.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_scorers = df_ball[df_ball['batsman'].isin(player_runs.index)]

# Plotting Run Distributions per Player
plt.figure(figsize=(10, 6))
sns.histplot(top_scorers, x='batsman_runs', hue='batsman', element='step', stat='count', common_norm=False, kde=True)
plt.title('Run Distributions per Top 10 Player')
plt.xlabel('Runs Scored in a Ball')
plt.ylabel('Frequency')
plt.legend(title='Player')
plt.tight_layout()
# Saving the formatted chart as a PNG file
plt.savefig('chart3.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Data preparation for Wicket Distribution Among Bowlers
top_wicket_takers = df_ball[df_ball['is_wicket'] == 1]['bowler'].value_counts().head(10)

# Plotting Wicket Distribution Among Bowlers
plt.figure(figsize=(12, 6))
sns.barplot(x=top_wicket_takers.index, y=top_wicket_takers.values, palette='mako')
plt.title('Top Wicket Takers in IPL')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)
plt.tight_layout()
# Saving the formatted table as a PNG file
plt.savefig('chart4.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Data preparation for Average Match Scores Over the Years
# Adding total runs per match in matches_data for easier calculation
total_runs_per_match = df_ball.groupby('id')['total_runs'].sum()
matches_data = df_matches.merge(total_runs_per_match, on='id', how='left')

# Calculating average scores per year
matches_data['year'] = pd.to_datetime(matches_data['date']).dt.year
average_scores_per_year = matches_data.groupby('year')['total_runs'].mean()

# Plotting Average Match Scores Over the Years
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_scores_per_year.index, y=average_scores_per_year.values, marker='o', color='b')
plt.title('Average Match Scores Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Score per Match')
plt.grid(True)
plt.tight_layout()
# Saving the formatted table as a PNG file
plt.savefig('chart5.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Data preparation for Impact of Winning Toss on Winning Matches
df_matches['toss_winner_won'] = (df_matches['toss_winner'] == df_matches['winner'])
toss_impact_counts = df_matches['toss_winner_won'].value_counts()

# Plotting Impact of Winning Toss on Winning Matches
plt.figure(figsize=(8, 8))
plt.pie(toss_impact_counts, labels=['Toss Winner Won', 'Toss Winner Lost'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Impact of Winning Toss on Winning Matches')
plt.tight_layout()
# Saving the formatted table as a PNG file
plt.savefig('chart6.png', bbox_inches='tight', dpi=200)
plt.show()
```

```python
# Prepare the relevant data for 'result_margin' from matches
matches_data = df_matches['result_margin'].dropna()
matches_stats = matches_data.describe()
matches_stats.loc['variance'] = matches_data.var()
matches_stats.loc['skewness'] = matches_data.skew()

# Aggregate total runs scored by each batsman
total_batsman_runs = df_ball.groupby('batsman')['batsman_runs'].sum()
batsman_runs_stats = total_batsman_runs.describe()
batsman_runs_stats.loc['variance'] = total_batsman_runs.var()
batsman_runs_stats.loc['skewness'] = total_batsman_runs.skew()

# Aggregate total extra runs given by each bowler
total_extra_runs = df_ball.groupby('bowler')['extra_runs'].sum()
extra_runs_stats = total_extra_runs.describe()
extra_runs_stats.loc['variance'] = total_extra_runs.var()
extra_runs_stats.loc['skewness'] = total_extra_runs.skew()

# Combine stats into a single DataFrame for matches, batsman runs, and extra runs
combined_stats = pd.concat([matches_stats, batsman_runs_stats, extra_runs_stats], axis=1)
combined_stats.columns = ['Result Margin', 'Batsman Runs', 'Extra Runs']

# Format the numerical values to two decimal places
formatted_stats = combined_stats.applymap(lambda x: f"{x:.2f}" if isinstance(x, float) else x)

# Define row labels
row_labels = combined_stats.index.tolist()

# Create the table plot for a snapshot of the data
fig, ax = plt.subplots(figsize=(8, 3))
ax.axis('tight')
ax.axis('off')
table_data = ax.table(cellText=formatted_stats.values, colLabels=formatted_stats.columns, rowLabels=row_labels, loc='center')
table_data.auto_set_font_size(False)
table_data.set_fontsize(10)
table_data.scale(1.2, 1.2)  # Adjust scale to fit your needs
# Saving the formatted table as a PNG file
plt.savefig('table.png', bbox_inches='tight', dpi=200)
plt.show()
```
